## 1. EXPLICACIÓN TEÓRICA Y CONCEPTUAL

Para diseñar un sistema de Inteligencia Artificial moderno, debemos dominar el vocabulario técnico y los fundamentos científicos que rigen el Procesamiento del Lenguaje Natural (NLP). Vamos a desglosar cada concepto clave del pipeline:

### A. ¿Qué es un Corpus?

En lingüística computacional y NLP, un **corpus** (del latín *cuerpo*) es un conjunto amplio, estructurado y digitalizado de textos o datos lingüísticos que se utiliza para el análisis estadístico, el entrenamiento de modelos o la extracción de conocimiento. En nuestro proyecto, el archivo `ventas.csv` —una vez procesado y convertido a texto— se convertirá formalmente en nuestro corpus de conocimiento.

### B. ¿Qué es NLP (Natural Language Processing)?

El **Procesamiento del Lenguaje Natural** es la rama de la Inteligencia Artificial que se ocupa de la interacción entre las computadoras y el lenguaje humano. Su objetivo es permitir que las máquinas entiendan, interpreten, manipulen y generen lenguaje de una manera que sea valiosa. NLP combina lingüística computacional con modelos estadísticos, de Machine Learning y Deep Learning.

### C. ¿Qué es RAG (Retrieval-Augmented Generation)?

**RAG** (Generación Aumentada por Recuperación) es una arquitectura de diseño de IA que optimiza la salida de un Modelo de Lenguaje Grande (LLM) haciendo referencia a una fuente de conocimiento externa y autoritativa antes de generar una respuesta.

En lugar de confiar únicamente en lo que el modelo aprendió durante su entrenamiento general, el sistema RAG busca información relevante en nuestro corpus (fase de *Retrieval*) y se la entrega al LLM junto con la pregunta del usuario para que este redacte una respuesta precisa (fase de *Generation*).

### D. ¿Qué son los Embeddings?

Un **Embedding** (incrustación) es una representación matemática en forma de vector (una lista de números decimales de alta dimensionalidad) que captura el **significado semántico** de una palabra, frase o documento.
A diferencia de los métodos antiguos que solo contaban palabras (como buscar la palabra exacta "auto"), los embeddings permiten que la computadora entienda que las palabras "vehículo", "coche" y "automóvil" están íntimamente relacionadas en significado, ubicándolas en coordenadas espaciales muy cercanas dentro de un mapa matemático.

### E. ¿Qué es una Base Vectorial (Vector Store)?

Una **Base Vectorial** es una base de datos diseñada específicamente para almacenar y buscar eficientemente vectores de embeddings de alta dimensionalidad. A diferencia de una base de datos relacional (como SQL, que busca por filas y columnas exactas), una base vectorial indexa los datos matemáticos y realiza búsquedas de "vecinos más cercanos" utilizando métricas como la *similitud coseno* o la *distancia euclidiana*. Esto permite encontrar qué fragmentos de texto se parecen más a la pregunta del usuario en milisegundos.

### F. ¿Qué es un Agente IA?

Un **Agente de Inteligencia Artificial** es un sistema de software autónomo que utiliza un LLM como su "motor de razonamiento" principal. A diferencia de un chatbot secuencial o un pipeline RAG simple (que solo sigue un camino predeterminado), un Agente puede:

1. Evaluar una situación o pregunta compleja.
2. Crear un plan de acción.
3. Seleccionar y ejecutar **herramientas** externas (ej. consultar una base de datos, ejecutar código Python o buscar en la web).
4. Observar el resultado de la herramienta y decidir si la respuesta es suficiente o si requiere ejecutar otra acción.

### G. ¿Qué hace LangChain?

**LangChain** es un framework de desarrollo de código abierto diseñado específicamente para simplificar la creación de aplicaciones impulsadas por modelos de lenguaje. Actúa como el "pegamento orquestador" del sistema. Proporciona abstracciones modulares para encadenar componentes: la carga de documentos, la división de textos, la conexión con bases vectoriales, la gestión de la memoria del chat y la creación de agentes autónomos.

### H. ¿Qué es Mistral AI?

**Mistral AI** es una empresa de inteligencia artificial enfocada en crear modelos de lenguaje grandes (LLMs) de código abierto y comerciales de alta eficiencia. Sus modelos (como *Mistral-7B*, *Mixtral* o *Mistral Large*) destacan por su excelente balance entre tamaño computacional, velocidad de respuesta, ventana de contexto y capacidades de razonamiento lógico, siendo ideales para integrarse en arquitecturas RAG corporativas o académicas.

## 2. DIFERENCIAS COMPRENSIVAS Y ARQUITECTURA

Para un arquitecto de soluciones, es fundamental responder a dos preguntas críticas de diseño que justifican la existencia de este proyecto:

### Diferencia entre un Modelo Base y un Sistema RAG

* **Modelo Base (Vanila LLM):** Es el modelo de lenguaje puro tal como sale de su entrenamiento (ej. Mistral AI interactuando directamente contigo). Su conocimiento está congelado en el tiempo (su fecha de corte de entrenamiento) y no sabe absolutamente nada sobre los datos privados de tu empresa o tu archivo `ventas.csv`. Si le preguntas sobre ellos, inventará datos (alucinación) o dirá que no los conoce.
* **Sistema RAG:** Es una arquitectura que envuelve al modelo base. No modifica el cerebro del LLM; en su lugar, actúa como un examen a "libro abierto". El sistema busca las respuestas correctas en el CSV y le dice al LLM: *"Lee este fragmento exacto de las ventas de la empresa y, basándote únicamente en esto, responde la pregunta del usuario"*.

### ¿Por qué un LLM no puede "memorizar" el CSV directamente?

Es común que los principiantes piensen: *"¿Por qué no copiamos todo el texto del CSV y se lo pegamos al prompt del LLM?"*. Esto es inviable en entornos reales por tres razones arquitectónicas:

1. **Límite de la Ventana de Contexto (Context Window):** Los LLMs tienen una capacidad máxima de tokens (palabras/caracteres) que pueden procesar simultáneamente en una sola consulta. Si tu dataset de ventas crece a miles de filas, superará instantáneamente este límite.
2. **Degradación de la Atención ("Lost in the Middle"):** Diversos estudios científicos demuestran que cuando alimentas a un LLM con contextos masivos de texto, su capacidad para extraer datos precisos disminuye drásticamente, ignorando información crítica situada en la mitad del texto provisto.
3. **Costo Financiero y de Cómputo:** Los proveedores de LLM cobran por la cantidad de tokens enviados en el prompt. Enviar todo el archivo CSV en cada pregunta del usuario resultaría económicamente prohibitivo y computacionalmente lento. RAG resuelve esto enviando únicamente las 3 o 4 filas más relevantes para la pregunta actual.

````python
# ==========================================

# SECCIÓN 3 — FASE 3: NORMALIZACIÓN Y LIMPIEZA

# ==========================================

## 1. EXPLICACIÓN TEÓRICA

### ¿Por qué se normaliza un dataset en un contexto de NLP y RAG?

En la ingeniería de datos convencional (como en bases de datos relacionales o business intelligence), la normalización busca reducir la redundancia y asegurar la integridad transaccional. Sin embargo, en el ámbito del **Procesamiento de Lenguaje Natural (NLP)** y **Sistemas RAG**, el propósito se transforma: **buscamos consistencia semántica y reducción del ruido sintáctico**.

Los modelos de lenguaje y los algoritmos de codificación de texto (Embeddings) son extremadamente sensibles a las variaciones gráficas. Si los datos crudos contienen inconsistencias, el rendimiento de todo el sistema de recuperación se degradará en las fases posteriores.

### ¿Qué problemas resuelve la normalización en esta etapa?

1. **Dispersión Vocálica e Inconsistencia de Formato:** Variaciones sutiles como `"USA"`, `"usa"`, `" USA "` o `"U.S.A."` representan exactamente la misma entidad geográfica. Para un ser humano es obvio, pero para un sistema informático son cadenas de caracteres (strings) completamente diferentes.
2. **Ambigüedad por Valores Nulos (NaNs):** La presencia de valores ausentes rompe la concatenación de texto cuando intentamos convertir filas de una tabla en "documentos narrativos" legibles por un LLM. Un valor `NaN` mal gestionado inyectará la palabra literal `"nan"` dentro del documento conceptual, lo que confundirá al modelo confundiéndola con ruido o terminología errónea.
3. **Anomalías Temporales:** Las fechas expresadas en formatos heterogéneos (ej. `2003-05-07` vs `5/7/2003`) impiden realizar filtros cronológicos precisos o que el modelo entienda la secuencialidad real de los eventos de ventas.

### ¿Cómo afecta la normalización al Retrieval (Recuperación)?

El *Retrieval* se basa en encontrar coincidencias exactas (búsquedas léxicas) o de significado (búsquedas semánticas). Si el usuario pregunta por `"Ventas en el estado de Nueva York"` y en nuestro dataset el estado está guardado aleatoriamente como `"NY"`, `"ny"` o `"New York"`, el componente de recuperación fallará en agrupar los registros correctos. Al estandarizar los datos, garantizamos que el recuperador acceda a un corpus homogéneo, aumentando drásticamente la **Exhaustividad (Recall)** y la **Precisión (Precision)** del sistema.

### ¿Cómo afecta a los Embeddings y a las Búsquedas Semánticas?

Los modelos de *Embeddings* (como los de *Sentence Transformers*) convierten el texto en vectores densos en un espacio multidimensional.

* **Ruido Sintáctico:** Espacios en blanco adicionales (`"  Motorcycles "`), tabulaciones o caracteres especiales extraños alteran el cálculo matemático del vector resultante. El modelo gastará "capacidad de atención" en procesar esos espacios vacíos en lugar de concentrarse en el significado real de las palabras.
* **Distorsión del Espacio Vectorial:** Dos filas idénticas, una limpia y otra con espacios y minúsculas desordenadas, generarán vectores con una distancia coseno distante entre sí, rompiendo la lógica de similitud semántica. La normalización asegura que textos con el mismo significado semántico aterricen en coordenadas vectoriales muy cercanas.

---

## 2. EXPLICACIÓN CONCEPTUAL: EL FLUJO DE LA INFORMACIÓN

Antes de ejecutar el código, es vital entender conceptualmente cómo se transforman los datos en esta fase:

```
[ Dataset Crudo (.csv) ]
       │
       ▼
[ 1. Limpieza de Espacios ] ──► Elimina espacios fantasmas al inicio/final de los strings.
       │
       ▼
[ 2. Manejo de Nulos ]      ──► Reemplaza NaNs por placeholders semánticos ("No Registrado").
       │
       ▼
[ 3. Casteo de Tipos ]      ──► Transforma texto plano a Datetime y Numéricos reales.
       │
       ▼
[ 4. Estandarización Textual]──► Convierte categorías a mayúsculas/minúsculas homogéneas.
       │
       ▼
[ Dataset Normalizado ]     ──► Listo para ser transformado en Documentos de Texto (FASE 4).

```

* **Qué entra:** Un DataFrame de Pandas con texto desordenado, espacios en blanco invisibles, fechas en formato *string* y valores faltantes (`NaN`).
* **Qué sale:** Un DataFrame estructurado, limpio y predecible, donde cada celda sigue una regla de formato estricta.
* **Qué problema resuelve:** Evita que el sistema RAG genere alucinaciones o sufra de pérdidas de información debido a formatos inconsistentes en el origen de datos.

---

## 3. CÓDIGO COMENTADO (Diseñado para Google Colab)

A continuación se presenta el bloque de código modular y educativo. Este código toma el dataset de muestra (`sales_data_sample.csv`), aplicando técnicas reproducibles y seguras en Colab.

```python
# =====================================================================
# LIBRERÍAS REQUERIDAS Y SU JUSTIFICACIÓN
# =====================================================================
# pandas: Librería estándar de Python para manipulación de estructuras de datos tabulares (DataFrames).
# numpy: Utilizada aquí específicamente para identificar y gestionar valores nulos (np.nan).
import pandas as pd
import numpy as np

def normalizar_dataset(path_entrada, encoding_detectado='unicode_escape'):
    """
    Función académica para normalizar y limpiar el dataset de ventas.

    Entrada: Ruta del archivo CSV crudo.
    Salida: DataFrame de Pandas completamente limpio y estandarizado.
    """
    print("[INFO] Iniciando el proceso de normalización y limpieza...")

    # -----------------------------------------------------------------
    # ETAPA 1: Carga segura del archivo
    # -----------------------------------------------------------------
    # Usamos 'unicode_escape' ya que los archivos de ventas suelen contener
    # caracteres especiales o codificaciones occidentales que rompen el UTF-8 estándar.
    df = pd.read_csv(path_entrada, encoding=encoding_detectado)
    print(f"[INFO] Dataset cargado. Dimensiones iniciales: {df.shape[0]} filas, {df.shape[1]} columnas.")

    # -----------------------------------------------------------------
    # ETAPA 2: Limpieza de los nombres de las columnas
    # -----------------------------------------------------------------
    # Un error común es tener espacios ocultos en los encabezados (ej: "SALES ").
    # .str.strip() elimina espacios al principio y al final.
    # .str.upper() fuerza a que todos los nombres estén en mayúsculas para evitar errores de tipeo.
    df.columns = df.columns.str.strip().str.upper()
    print("[INFO] Columnas normalizadas a mayúsculas y sin espacios residuales.")

    # -----------------------------------------------------------------
    # ETAPA 3: Normalización de Fechas (Temporal)
    # -----------------------------------------------------------------
    # Por defecto, Pandas lee las fechas como texto (object).
    # Al pasar RAG a un CSV, el formato de fecha debe ser homogéneo (YYYY-MM-DD).
    # errors='coerce' transforma cualquier fecha corrupta o inválida en un valor nulo (NaT) en lugar de romper el código.
    if 'ORDERDATE' in df.columns:
        df['ORDERDATE'] = pd.to_datetime(df['ORDERDATE'], errors='coerce')
        print("[INFO] Columna 'ORDERDATE' convertida exitosamente a formato Datetime estándar.")

    # -----------------------------------------------------------------
    # ETAPA 4: Limpieza de Espacios en Columnas de Texto (Sintaxis)
    # -----------------------------------------------------------------
    # Identificamos columnas de tipo objeto (texto) para aplicarles limpieza de strings.
    columnas_texto = df.select_dtypes(include=['object']).columns

    for col in columnas_texto:
        # .str.strip() elimina espacios en blanco al inicio y al final de cada celda de texto.
        df[col] = df[col].astype(str).str.strip()

    print(f"[INFO] Espacios en blanco eliminados en {len(columnas_texto)} columnas de texto.")

    # -----------------------------------------------------------------
    # ETAPA 5: Manejo de Valores Nulos (Semántica para RAG)
    # -----------------------------------------------------------------
    # En sistemas RAG tabulares, si dejamos campos vacíos, el LLM no sabrá cómo interpretarlos.
    # Rellenamos los vacíos con un string explícito que denote la ausencia de datos.
    columnas_criticas_nulos = ['ADDRESSLINE2', 'STATE', 'POSTALCODE', 'TERRITORY']

    for col in columnas_criticas_nulos:
        if col in df.columns:
            # Reemplazamos valores vacíos, strings "nan" o nulos reales por "No Registrado"
            df[col] = df[col].replace(['nan', 'NaN', '', None], 'No Registrado')
            # Aseguramos el llenado de nulos remanentes nativos de Pandas
            df[col] = df[col].fillna('No Registrado')

    print("[INFO] Valores nulos mitigados y reemplazados por placeholders semánticos.")

    # -----------------------------------------------------------------
    # ETAPA 6: Estandarización Textual de Categorías
    # -----------------------------------------------------------------
    # Para facilitar las búsquedas semánticas y agrupaciones del LLM,
    # estandarizamos las columnas de categorías clave a Mayúsculas Sostenidas.
    columnas_categoricas = ['STATUS', 'PRODUCTLINE', 'DEALSIZE', 'COUNTRY']

    for col in columnas_categoricas:
        if col in df.columns:
            df[col] = df[col].str.upper()

    print("[INFO] Columnas categóricas estandarizadas a MAYÚSCULAS.")

    # -----------------------------------------------------------------
    # ETAPA 7: Verificación Final de Consistencia
    # -----------------------------------------------------------------
    # Comprobamos que no queden nulos críticos que puedan romper la fase de embeddings.
    nulos_restantes = df.isnull().sum().sum()
    print(f"[INFO] Proceso concluido. Valores nulos remanentes en el dataset: {nulos_restantes}")

    return df

# =====================================================================
# EJECUCIÓN EN EL ENTORNO DE GOOGLE COLAB
# =====================================================================
try:
    df_limpio = normalizar_dataset('sales_data_sample.csv')

    # Guardamos el archivo limpio para que pueda ser descargado o usado directamente
    archivo_salida = 'ventas_normalizado.csv'
    df_limpio.to_csv(archivo_salida, index=False)
    print(f"\n[ÉXITO] Archivo listo para Google Colab guardado como: '{archivo_salida}'")

    # Inspección rápida de los resultados de la transformación
    print("\n--- MUESTRA DE DATOS NORMALIZADOS ---")
    print(df_limpio[['ORDERDATE', 'STATUS', 'PRODUCTLINE', 'STATE', 'COUNTRY']].head())

except FileNotFoundError:
    print("[ERROR] No se encontró el archivo 'sales_data_sample.csv'. Por favor, súbelo al panel izquierdo de Colab.")

```
````
